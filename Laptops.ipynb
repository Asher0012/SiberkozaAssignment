```python
df = pd.read_csv('Laptops.csv')
df.head()
```

```python
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df.head()
```

```python
df['Price'] = df['Price'].str.replace(',', '')
df['Price'] = df['Price'].str.replace('₹', '')
df.head(5)
```

```python
df["RAM"] = df.RAM.str.replace('\s.*','')

df.head(5)
```

```python
df["Warranty"] = df.Warranty.str.replace('\s.*','')
df.head()
```

```python
df.drop(['rating', 'Operating System'], axis = 1, inplace = True)
df.head()
```

```python
df["Price"] = df["Price"].astype(str).astype(int)
print(df.dtypes)
```

```python
df['Storage'] = df['Storage'].replace('1 TB', '1024 GB', regex=True)
df.head(5)
#Changing Storage to GB
```

```python
column = [Predictor for Predictor in df.columns if df[Predictor].dtype == 'O']
for Predictor in column:
    labels_ordered= df.groupby([Predictor])['Price'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[Predictor]=df[Predictor].map(labels_ordered)
df.head(10)
```

```python
Target = df['Price']
features = df.drop(columns='Price')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 

a = scaler.fit_transform(features)
features = pd.DataFrame(a)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, Target, test_size=0.3, random_state=42,shuffle = True)
#Splitting data for testing and training
```

```python
from sklearn.ensemble import RandomForestRegressor
Random_forest_regressor = RandomForestRegressor(random_state=10)
Random_forest_regressor.fit(X_train,y_train)
#using Random forest Regressor
```

```python
print("Training score")
print(Random_forest_regressor.score(X_train,y_train))
print('\r')
print("Testing score")
print(Random_forest_regressor.score(X_test,y_test))
```
